```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def delete_zero_sum_nodes(head):
    dummy = Node(0)
    dummy.next = head
    prefix_sum = 0
    node_map = {}
    current = dummy

    while current:
        prefix_sum += current.data
        if prefix_sum in node_map:
            node_map[prefix_sum].next = current.next
        else:
            node_map[prefix_sum] = current
        current = current.next

    return dummy.next


def reverse_linked_list_in_groups(head, k):
    current = head
    next_node = None
    prev = None
    count = 0

    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    if next_node:
        head.next = reverse_linked_list_in_groups(next_node, k)

    return prev


def merge_linked_lists(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    p = head1
    q = head2

    while p and q:
        temp1 = p.next
        temp2 = q.next

        q.next = temp1
        p.next = q

        p = temp1
        q = temp2

    return head1


def count_pairs_with_sum(arr, target_sum):
    count = 0
    num_map = {}

    for num in arr:
        complement = target_sum - num
        if complement in num_map:
            count += num_map[complement]
        if num in num_map:
            num_map[num] += 1
        else:
            num_map[num] = 1

    return count


def find_duplicates(arr):
    duplicates = []
    num_map = {}

    for num in arr:
        if num in num_map:
            num_map[num] += 1
        else:
            num_map[num] = 1

    for num, count in num_map.items():
        if count > 1:
            duplicates.append(num)

    return duplicates


def find_kth_largest_smallest(arr, k):
    arr.sort()
    kth_largest = arr[-k]
    kth_smallest = arr[k - 1]
    return kth_largest, kth_smallest


def move_negative_elements(arr):
    i = 0
    j = len(arr) - 1

    while i <= j:
        if arr[i] < 0 and arr[j] >= 0:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
            j -= 1
        elif arr[i] >= 0:
            i += 1
        else:
            j -= 1

    return arr


def reverse_string_using_stack(string):
    stack = []
    for char in string:
        stack.append(char)

    reversed_string = ""
    while stack:
        reversed_string += stack.pop()

    return reversed_string


def evaluate_postfix_expression(expression):
    stack = []

    for char in expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand2 = stack.pop()
            operand1 = stack.pop()
            result = perform_operation(char, operand1, operand2)
            stack.append(result)

    return stack.pop()


def perform_operation(operator, operand1, operand2):
    if operator == "+":
        return operand1 + operand2
    elif operator == "-":
        return operand1 - operand2
    elif operator == "*":
        return operand1 * operand2
    elif operator == "/":
        return operand1 / operand2


class QueueUsingStack:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, data):
        self.stack1.append(data)

    def dequeue(self):
        if not self.stack2:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        if self.stack2:
            return self.stack2.pop()
        else:
            return None
```
